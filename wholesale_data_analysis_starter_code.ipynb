## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty

df['subtotal'] = df['unit_price'] * df['qty']
df.head()

```

```python
# Create a column for shipping price. Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.

# Calculate the total weight for each order line
df['total_weight'] = df['unit_weight'] * df['qty']
# df.head()

# Create a new column 'shipping_price' based on the total weight
df['shipping_price'] = df['total_weight'].apply(lambda x: 7 * x if x > 50 else 10 * x)
df.head()
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25%

# Create a new column 'subtotal' by multiplying 'unit_price' and 'qty'
df['subtotal'] = df['unit_price'] * df['qty']

# Calculate the total price including sales tax
df['total_price'] = (df['subtotal'] + df['shipping_price']) * 1.0925

# Display the first few rows to verify the new columns
df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and shipping price (assume the shipping cost is exactly what is charged to the client).

# Calculate the cost for each line
df['line_cost'] = (df['unit_cost'] * df['qty']) + df['shipping_price']
df.head()

```

```python
# Create a column for the profit of each line using line cost and line price

# Calculate the profit for each line
df['profit'] = df['total_price'] - df['line_cost']
df.head()
```

Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids.
# define a function that will revieve the client id and for loop grap the line price

top_5_data = df[df['client_id'].isin(top_five_client_ids)]
top_5_data
```

```python
## Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
## total units purchased, total shipping price, total revenue, and total profit. 

client_counts = df['client_id'].value_counts()
top_five_clients = client_counts.head(5)
top_five_client_ids = top_five_clients.index.tolist()

top_5_data = df[df['client_id'].isin(top_five_client_ids)]



# Calculate the required metrics for the top 5 clients
# define a function and for loop
summary = top_5_data.groupby('client_id').agg(
    total_units_purchased=('qty', 'sum'),
    total_shipping_price=('shipping_price', 'sum'),
    total_revenue=('total_price', 'sum'),
    total_profit=('profit', 'sum')
).reset_index()

print(summary)
```

```python
# Format the data and rename the columns to names suitable for presentation.

display_summary_df = summary.rename(columns={
    'client_id': 'Client ID',
    'total_units_purchase': 'Units',
    'total_shipping_price': 'Shipping Price',
    'total_revenue': 'Total Revenue',
    'total_profit': 'Total Profit'
})

print(display_summary_df)


# Define a function that converts a dollar amount to millions.
def scale_to_millions(amount):
    return amount / 1000000
```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.

sorted_data = display_summary_df.sort_values(by='Total Profit', ascending=False)
sorted_data
```
